```
from sentence_transformers import SentenceTransformer, models
import pandas as pd
from torch import nn
```

```
word_embedding_model = models.Transformer('Musixmatch/umberto-commoncrawl-cased-v1')
pooling_model = models.Pooling(word_embedding_model.get_word_embedding_dimension(),
                               pooling_mode_mean_tokens=True,
                               pooling_mode_cls_token=True,
                               pooling_mode_max_tokens=True)
dense_model = models.Dense(in_features=pooling_model.get_sentence_embedding_dimension(), out_features=768, activation_function=nn.Tanh())

model = SentenceTransformer(modules=[word_embedding_model, pooling_model, dense_model])
```

```
sentences = list(pd.read_csv('dankmemes_task1_train.csv')['Text'].values)
sentence_embeddings = model.encode(sentences)
```

```
import pickle
import os

pickle.dump(sentence_embeddings,
            open(os.path.join('umberto_embeddings.pkl'), 'wb'),
            protocol=4)
```
