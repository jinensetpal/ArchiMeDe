```
from img2vec_pytorch import Img2Vec
from PIL import Image

img2vec = Img2Vec(cuda=True, model='alexnet')

img = Image.open('dbscan.png').convert('RGB')
vec = img2vec.get_vec(img, tensor=False)
```

```
img = Image.open('127.jpg').convert('RGB')
vec = img2vec.get_vec(img, tensor=False)
alexnet.append(vec)
```

```
for i in range(128, 1600):
  img = Image.open('drive/My Drive/Colab Notebooks/images/' + str(i) + '.jpg').convert('RGB')
  vec = img2vec.get_vec(img, tensor=False)
  print('Vectorized: ' + str(i) + '.jpg')
  alexnet.append(vec)
```

```
import numpy as np

alex = np.array(alexnet)
```

```
import os
import pickle

pickle.dump(alex,
            open(os.path.join('alexnet_embeddings.pkl'), 'wb'),
            protocol=4)

! cp alexnet_embeddings.pkl ./drive/My\ Drive/Colab\ Notebooks/
```
