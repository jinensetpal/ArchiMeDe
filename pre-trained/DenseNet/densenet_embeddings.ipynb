```
embeddings = []
for i in range(1600):
  img = Image.open('drive/My Drive/Colab Notebooks/images/' + str(i) + '.jpg').convert('RGB')
  transform = transforms.Compose([transforms.ToTensor()])
  img_t = transform(img)
  batch_t = torch.unsqueeze(img_t, 0)
  vec = densenet(batch_t).detach().numpy().ravel()
  print('Vectorized: ' + str(i) + '.jpg')
  embeddings.append(vec)
```

```
import numpy as np

dense = np.array(embeddings)
```

```
import os
import pickle

pickle.dump(dense,
            open(os.path.join('densenest_embeddings.pkl'), 'wb'),
            protocol=4)

! cp densenest_embeddings.pkl ./drive/My\ Drive/Colab\ Notebooks/
```
